## Logistic Regression Workshop

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

df=pd.read_csv('iris-data-clean.csv')

df.head()
```

```python

encode=LabelEncoder()
df['class']=encode.fit_transform(df['class'])
df.head()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

logistic=LogisticRegression()
x=df[['sepal_length_cm','sepal_width_cm','petal_length_cm','petal_width_cm']]
y=df['class']


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
logistic.fit(x_train, y_train)

y_pre=logistic.predict(x_test)
accuracy_score(y_test,y_pre)
print(confusion_matrix(y_test,y_pre))
```

```python
x_sepal_length=df[['sepal_length_cm']]
x_sepal_width=df[['sepal_width_cm']]
logistic_length = LogisticRegression()
logistic_width = LogisticRegression()

logistic_length.fit(x_sepal_length, y)
logistic_width.fit(x_sepal_width, y)

y_pre_length = logistic_length.predict(x_sepal_length)
y_pre_width = logistic_width.predict(x_sepal_width)

print("Accuracy using sepal length only:", accuracy_score(y, y_pre_length))
print("Accuracy using sepal width only:", accuracy_score(y, y_pre_width))
```
